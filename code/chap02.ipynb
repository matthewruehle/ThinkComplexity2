## Complete graph

To make a complete graph, we use a generator function that iterates through all pairs of nodes.

```python
def all_pairs(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i < j:
                yield u, v
```

`make_complete_graph` makes a `Graph` with the given number of nodes and edges between all pairs of nodes.

```python
def make_complete_graph(n):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(all_pairs(nodes))
    return G
```

Here's a complete graph with 10 nodes:

```python
complete = make_complete_graph(10)
len(complete.nodes())
```

And here's what it looks like.

```python
nx.draw_circular(complete, 
                 node_color=COLORS[2], 
                 node_size=1000, 
                 with_labels=True)
plt.savefig('chap02-3.pdf')
```

The `neighbors` method returns a list of neighbors for a given node.

```python
complete.neighbors(0)
```

## Random graphs

Next we'll make a random graph where the probability of an edge between each pair of nodes is $p$.

The helper function `flip` returns True with probability `p` and False with probability `1-p`

```python
from numpy.random import random

def flip(p):
    return random() < p
```

`random_pairs` is a generator function that enumerates all possible pairs of nodes and yields each one with probability `p` 

```python
def random_pairs(nodes, p):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i<j and flip(p):
                yield u, v
```

`make_random_graph` makes an ER graph where the probability of an edge between each pair of nodes is `p`.

```python
def make_random_graph(n, p):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(random_pairs(nodes, p))
    return G
```

Here's an example with `n=10` and `p=0.3`

```python
random_graph = make_random_graph(10, 0.3)
len(random_graph.edges())
```

And here's what it looks like:

```python
nx.draw_circular(random_graph, 
                 node_color=COLORS[3], 
                 node_size=1000, 
                 with_labels=True)
plt.savefig('chap02-4.pdf')
```

## Connectivity

To check whether a graph is connected, we'll start by finding all nodes that can be reached, starting with a given node:

```python
def reachable_nodes(G, start):
    seen = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in seen:
            seen.add(node)
            stack.extend(G.neighbors(node))
    return seen
```

In the complete graph, starting from node 0, we can reach all nodes:

```python
reachable_nodes(complete, 0)
```

In the random graph we generated, we can also reach all nodes (but that's not always true):

```python
reachable_nodes(random_graph, 0)
```

We can use `reachable_nodes` to check whether a graph is connected:

```python
def is_connected(G):
    start = next(G.nodes_iter())
    reachable = reachable_nodes(G, start)
    return len(reachable) == len(G)
```

Again, the complete graph is connected:

```python
is_connected(complete)
```

But if we generate a random graph with a low value of `p`, it's not:

```python
random_graph = make_random_graph(10, 0.1)
len(random_graph.edges())
```

```python
is_connected(random_graph)
```

**Exercise:** What do you think it means for a directed graph to be connected?  Write a function that checks whether a directed graph is connected.

```python
def reachable_nodes_directed(G, start):
    # Assumes, or disregards, self-connectivity. 
    # This could also require self-connectedness, but with the existing random generator... \_()_/
    seen = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in seen:
            seen.add(node)
            stack.extend(G.successors(node))
    return seen
    
def is_directed_strong_connected(G):
    start = next(G.nodes_iter())
    reachable = reachable_nodes_directed(G, start)
    return (len(reachable) == len(G))

def random_directional_pairs(nodes, p):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if flip(p):
                yield u, v

def make_random_digraph(n, p):
    G = nx.DiGraph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(random_directional_pairs(nodes, p))
    return G

random_digraph = make_random_digraph(10, 0.25)

nx.draw_circular(random_digraph, 
                 node_color=COLORS[3], 
                 node_size=1000, 
                 with_labels=True)

is_directed_strong_connected(random_digraph)
```

## Probability of connectivity

Now let's estimare the probability that a randomly-generated ER graph is connected.

This function takes `n` and `p`, generates `iters` graphs, and returns the fraction of them that are connected.

```python
def prob_connected(n, p, iters=100):
    count = 0
    for i in range(iters):
        random_graph = make_random_graph(n, p)
        if is_connected(random_graph):
            count += 1
    return count/iters
```

With `n=10` and `p=0.3`, the probability of being connected is about 65%.

```python
n = 10
prob_connected(n, 0.3, iters=10000)
```

According to Erdos and Renyi, the critical value of `p` for `n=10` is about 0.23. 

```python
pstar = np.log(n) / n
pstar
```

So let's plot the probability of connectivity for a range of values for `p`

```python
ps = np.logspace(-1.3, 0, 11)
ps
```

I'll estimate the probabilities with `iters=1000`

```python
ys = [prob_connected(n, p, 1000) for p in ps]

for p, y in zip(ps, ys):
    print(p, y)
```

And then plot them, adding a vertical line at the computed critical value

```python
import thinkplot

thinkplot.vlines([pstar], 0, 1, color='gray')
thinkplot.plot(ps, ys)
thinkplot.config(xlabel='p', ylabel='prob connected', xscale='log', xlim=[ps[0], ps[-1]])
plt.savefig('chap02-5.pdf')
```

We can run the same analysis for a few more values of `n`.

```python
ns = [30, 100, 300]
ps = np.logspace(-2.5, 0, 11)

thinkplot.preplot(len(ns))
for n in ns:
    pstar = np.log(n) / n
    thinkplot.vlines([pstar], 0, 1, color='gray')

    ys = [prob_connected(n, p) for p in ps]
    thinkplot.plot(ps, ys, label='n=%d' % n)

thinkplot.config(xlabel='p', ylabel='prob connected',
                 xscale='log', xlim=[ps[0], ps[-1]],
                 loc='upper left')
plt.savefig('chap02-6.pdf')
```

As `n` increases, the critical value gets smaller and the transition gets more abrupt.

    def is_connected(G):

When checking the connectivity of a graph, we calculate len(G), which will not increase the order of O(n+m). 
(as it is either a linear or constant-time call), we grab a "starting" node, and we calculate the length of a list.
As a result, the order of growth for is_connected is essentially that of reachable_nodes, plus operations which
will just affect the coefficients or add constant time.
The order of growth for is_connected thus remains O(n+m).

**Exercise:** In my implementation of `reachable_nodes`, you might be bothered by the apparent inefficiency of adding *all* neighbors to the stack without checking whether they are already in `seen`.  Write a version of this function that checks the neighbors before adding them to the stack.  Does this "optimization" change the order of growth?  Does it make the function faster?

```python
def reachable_nodes_precheck(G, start):
    seen = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in seen:
            seen.add(node)
            stack.extend([i for i in G.neighbors(node) if node not in seen])
    return seen
```

This optimization does not appear to change the order of growth. Checking the neighbors in advance merely changes when the "not in seen" check occurs, and decreases how many nodes need to be added (via extend) and removed (via pop). Appending and pop are constant-time, and thus the order of growth should be unaffected; the precheck version should be faster but not by a different order.

```python
%timeit len(reachable_nodes(complete, 0))
```

```python
%timeit len(reachable_nodes_precheck(complete, 0))
```
